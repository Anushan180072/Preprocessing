```python
import pandas as pd
import numpy as np
```

```python
fname="/content/drive/MyDrive/f1 - Copy.xlsx"
df = pd.read_excel(f"{fname}", keep_default_na=False, header=None)
```

```python
df
```

```python
df.drop(df.head(3).index, inplace=True)
df.drop(df.tail(2).index, inplace=True)
```

```python
df
```

```python
df.replace("", np.nan, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.dropna(how="all", axis=0, inplace=True)
```

```python
df
```

```python
df.reset_index(inplace=True)
df.columns = np.arange(0, len(df.columns))
```

```python
df
```

```python
df.dropna(subset=[df.columns[-1]], inplace=True)
```

```python
df
```

```python
#to remove total for text rows
df = df[~df.apply(lambda row: row.astype(str).str.contains("Total for").any(), axis=1)]
df
```

```python
#To remove repeated header
header_row = df.iloc[0]
df=df[df.apply(lambda row: not row.str.strip().equals(header_row.str.strip()), axis=1)]


```

```python
df
```

```python
df.to_csv(f"{fname}.csv", header=None, index=None)
```
