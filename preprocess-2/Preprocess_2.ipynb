```python
import pandas as pd
import numpy as np
```

```python
fname="/content/report (1).xlsx"
df = pd.read_excel(f"{fname}", keep_default_na=False, header=None)
```

```python
df
```

```python
df.drop(df.head(1).index, inplace=True)
df.drop(df.tail(1).index, inplace=True)
```

```python
df.replace("", np.nan, inplace=True)
df.dropna(how="all", axis=1, inplace=True)
df.dropna(how="all", axis=0, inplace=True)
```

```python
df = df.applymap(lambda x: x.replace('%', '') if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace('$', '') if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace('(', '') if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace(')', '') if isinstance(x, str) else x)
df
```

```python
df.reset_index(inplace=True)
df.columns = np.arange(0, len(df.columns))
df
```

```python
df[4]=df[4].fillna(df[3])
df[6]=df[6].fillna(df[5])
df[9]=df[9].fillna(df[7])
df[9]=df[9].fillna(df[8])
df[11]=df[11].fillna(df[10])
df[11]=df[11].fillna(df[12])
df[14]=df[14].fillna(df[13])
df.drop(columns=[3,5,7,8,10,12,13],inplace=True)
```

```python
df.iloc[3:4,2:]=df.iloc[2:3,2:]
df.iloc[2:3,2:]=np.nan

```

```python
df.columns = np.arange(0, len(df.columns))
```

```python
fname="/content/report"
df.to_csv(f"{fname}.csv", header=None, index=None)
```
